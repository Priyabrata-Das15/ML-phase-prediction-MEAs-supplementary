```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

ID=pd.read_excel('G:\Priyabrata\ML\Individual data.xlsx','MSS+SSS+SSIM')
print(ID)
```

```python
INPUT = ID.iloc[:,2:11]  #all INPUT (Variables)
INPUT

TARGET = ID.iloc[:,13]
TARGET

#Y = ID.iloc[:,14] #coded phases. 0:MSS, 1:SSS
#Y
```

```python
# Visualization in 2D . valid only when u have pc1 and pc2.
fig = plt.figure(figsize = (8,8),dpi=300)
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Mixing enthalpy ($\mathbf{\Delta}$$\mathbf{H_{mix}}$) (KJ/mol)',fontsize = 18, weight='bold')
ax.set_ylabel('Atomic size mismatch ($\mathbf{\delta}$)',fontsize = 18,weight='bold')
#change the plot axis limits
ax.set_xlim([-40,10]) 
ax.set_ylim([0,12])
ax.xaxis.set_tick_params(labelsize=16)
ax.yaxis.set_tick_params(labelsize=16)
#ax.set_title('Plot of Ω Vs Λ', fontsize = 20)

targets = ['SSS','MSS', 'SS+IM']

colors = ['orange', 'purple', 'brown']
markers = ["1","+","*"]

for target, marker in zip(targets,markers):
        indicesToKeep = TARGET == target
        ax.scatter(INPUT.loc[indicesToKeep, 'Mixing enthalpy']
               , INPUT.loc[indicesToKeep, 'Atomic size mismatch']
               , marker=marker, alpha= 1
               , s = 100)
    

    
ax.legend(targets,fontsize=16)

plt.rcParams["font.weight"]="bold"    
```
